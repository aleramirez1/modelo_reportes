# EDA_datset

Limpieza y análisis exploratorio del dataset recolecta

## Paso 1: Cargar y leer el dataset

```python
from pathlib import Path
import pandas as pd

base_dir = Path.cwd().parent if Path.cwd().name == 'notebook' else Path.cwd()
data_path = base_dir / 'data' / 'dataset_reportes.csv'

df = pd.read_csv(data_path)
print('Ruta:', data_path)
print('Shape:', df.shape)
display(df.head())
```

## Paso 2: Revisar tipos, nulos y duplicados

```python
print('Tipos de datos:')
display(df.dtypes.to_frame('dtype'))

print('\nNulos por columna:')
display(df.isnull().sum().to_frame('nulos'))

duplicados = int(df.duplicated().sum())
print(f'\nRegistros duplicados: {duplicados}')

resumen_paso2 = pd.DataFrame({
    'filas': [len(df)],
    'columnas': [df.shape[1]],
    'duplicados': [duplicados],
    'nulos_totales': [int(df.isnull().sum().sum())]
})

display(resumen_paso2)
```

## Paso 3: Ver datos faltantes

```python
faltantes = df.isnull().sum()
porcentaje_faltantes = (df.isnull().mean() * 100).round(2)

resumen_faltantes = pd.DataFrame({
    'faltantes': faltantes,
    'porcentaje_%': porcentaje_faltantes
}).sort_values('faltantes', ascending=False)

print('Resumen de datos faltantes por columna:')
display(resumen_faltantes)

filas_con_faltantes = int(df.isnull().any(axis=1).sum())
print(f'Filas con al menos un dato faltante: {filas_con_faltantes} de {len(df)}')

if filas_con_faltantes > 0:
    display(df[df.isnull().any(axis=1)].head())
```
